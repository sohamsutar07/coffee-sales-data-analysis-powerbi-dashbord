```python
coffee_sales.groupby("coffee_name")["money"].sum().plot(kind="bar")

plt.title("Coffee Sales by Product")

plt.xlabel("Coffee Name")

plt.ylabel("Total Sales")

plt.show()
```

```python
coffee_sales.groupby("cash_type")["money"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```
